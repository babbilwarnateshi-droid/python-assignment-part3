### *TASK 1 — FILE READ & WRITE BASICS*

```python
# ======================================================
# PART A - WRITE & APPEND
# ======================================================

file_name = "python_notes.txt"

# Initial content 
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]

try:
    #WRITE MODE -> overwrites file (fresh dataset creation)
    # Adding utf-8 encoding to handles potential charecters or emojis safely
    with open(file_name, "w", encoding = "utf-8") as file:
        for line in notes:
            file.write(line + "\n")     # newline ensures proper formatting.

    print("✅ File written successfully")

except Exception as e:
    print("❌ Error writting file:", e)


#APPEND MODE -> adds new data without deleting old content 
extra_notes = [
    "Topic 6: Functions improve code reusability.",
    "Topic 7: File handling enables data persistence."
]

try:
    with open(file_name, "a", encoding = "utf-8") as file:
        for line in extra_notes:
            file.write(line + "\n")
    
    print("✅ Lines appended successfully.")

except Exception as e:
    print("❌ Error appending file:", e)



# ======================================================
# PART B - READ & ANALYZE
# ======================================================

try:
    with open(file_name, "r", encoding = "utf-8") as file:
        lines = file.readlines()   #reads all lines into list
    
    print("\n====== FILE CONTENT ======")

    #Print numbered lines (cleaned output)
    for i, line in enumerate(lines, start = 1):
        clean_line = line.strip()  # removes \n and extra spaces
        print(f"{i}. {clean_line}")


    # Count total lines
    total_lines = len(lines)
    print(f"\nTotal Lines: {total_lines}")

except FileNotFoundError:
    print(f"❌ File not found. Please run write step first.")

except Exception as e:
    print("❌ Error reading file:", e)



# ======================================================
# KEYWORD SEARCH (CASE-INSENSITIVE)
# ======================================================

keyword = input("\nEnter keyword to search:").strip().lower()

found = False 

print("\n====== SEARCH RESULTS ======")


for line in lines:
    # Case-insensitive matching
    if keyword in line.lower():
        print(line.strip())
        found = True


# If no match found
if not found:
    print("⚠ No matching lines found.")
```

### *TASK 2 — API INTEGRATION*

```python
import requests

BASE_URL = "https://dummyjson.com/products"


# ======================================================
# STEP 1 - FETCH PRODUCTS
# ======================================================

print("\n====== FETCHING PRODUCTS ======")

try:
    response = requests.get(f"{BASE_URL}?limit=20", timeout = 10)

    # Check if request successful 
    response.raise_for_status()

    data = response.json()
    products = data.get("products", [])

    print("\nID  | Title                          | Category      | Price     | Rating")
    print("----|--------------------------------|---------------|-----------|---------")

    # Display formatted table 
    for p in products:
        print(f"{p['id']:<3} | {p['title'][:30]:<30} | {p['category']:<13} | ${p['price']:<8} | {p['rating']}")


except requests.exceptions.RequestException as e:
    print("❌ API request failed:", e)


# ======================================================
# STEP 2 - FILTER & SORT
# ======================================================

print("\n====== FILTERED (Rating ≥  4.5) & SORTED ======")

try:

    # Filter products based on rating
    filtered = [p for p in products if p.get ("rating", 0) >= 4.5]

    # Sort by price (descending)
    filtered_sorted = sorted(filtered, key = lambda x: x["price"], reverse = True)

    for p in filtered_sorted:
        print(f"{p['title']} -> ${p['price']} (Rating: {p['rating']})")

except Exception as e:
    print("❌ Error in filtering/sorting:", e)


# ======================================================
# STEP 3 - FETCH LAPTOPS CATEGORY
# ======================================================  

print("\n====== LAPTOPS CATEGORY ======")

try:

    response = requests.get(f"{BASE_URL}/category/laptops", timeout = 10)
    response.raise_for_status()

    laptops = response.json().get("products",[])

    for laptop in laptops:
        print(f"{laptop['title']} -> ${laptop['price']}")

except requests.exceptions.RequestException as e:
    print("❌ Failed to fetch laptops:", e)


# ======================================================
# STEP 4 - POST REQUEST
# ======================================================     

print("\n===== POST REQUEST (SIMULATION) ======")

new_product = {
    "title" : "My Custom Product",
    "price" : 999,
    "category" : "electronics",
    "description" : "A product I created via API"
}

try:

    response = requests.post(f"{BASE_URL}/add", json = new_product, timeout = 10)
    response.raise_for_status()

    result = response.json()

    print("Response from server:")
    print(result)

except requests.exceptions.RequestException as e:
    print("❌ POST request failed:", e) 
```

### *TASK 3 — EXCEPTION HANDLING*

```python
import requests


# ======================================================
# PART A - GUARDED CALCULATOR
# ======================================================   

def safe_divide(a,b):
    '''
    Performs safe divisions with error handling.
    Prevents runtime failures in calculations (financial models and KPIs) 
    '''

    try:
        return a / b
    # Using specific exception types like ZeroDivisionError instead of a 'catch-all' 
    # except block for better debugging
    except ZeroDivisionError:           
        return "Error: Cannot divide by zero"
    
    except TypeError:
        return "Error: Invalid input types"
    

    # TEST CASES
    print("\n====== SAFE DIVIDE TEST ======")
    print("10 / 2 =", safe_divide(10,2))
    print("10 / 0 =", safe_divide(10,0))
    print('"ten" / 2=', safe_divide("ten",2))



# ======================================================
# PART B - GUARDED FILE READER
# ====================================================== 

def read_file_safe(filename):
    '''
    Safely reads a file.
    Prevents system crash due to missing files (common in pipelines).
    '''

    try:
        with open(file_name, "r", encoding = "utf-8") as file:
            content = file.read()
            return content
        
    
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")


    finally:
        print("File operation attempt complete.")
    # The finally block is essential here 
    # to ensure this message prints regardless of success or failure.

    # TEST CASES
    print("\n====== FILE READ TESTS ======")

    print("\nReading python_notes.txt:")
    print(read_file_safe("python_notes.txt"))

    print("\nReadinf ghost_file.txt:")
    print(read_file_safe("ghost_file.txt"))


# ======================================================
# PART C - ROBUST API CALLS
# ======================================================     

print("\n====== ROBUST API CALL ======")

url = "https://dummyjson.com/products?limit=5"

try:
    response = requests.get(url, timeout = 5)
    response.raise_for_status()

    data = response.json()

    for p in data.get("products",[]):
        print(p["title"], "-",p["price"])


except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Unexpected error:",e)


# ======================================================
# PART D - INPUT VALIDATION LOOP
# ======================================================   

print("\n====== PRODUCT LOOKUP ======")

while True:
    user_input = input("\nEnter product ID (1-100) or 'quit': ").strip().lower()

    # Exit condition
    if user_input == 'quit':
        print("Exiting program.")
        break

    # VALIDATION: Check integer input to prevent the program from crashing on string input.
    try:
        product_id = int(user_input)

        # Range validation
        if not (1 <= product_id <= 100):
            print("⚠ Please enter a number between 1 and 100.")
            continue

    except ValueError:
        print("⚠ Invalid input. Enter a valid number.")
        continue

    
    # API CALL
    try: 
        response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout = 5)

        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            product = response.json()
            print(f"{product['title']} -> ${product['price']}")

        else:
            print("Unexpected response:", response.status_code)

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Unexpected error:", e)
```

### *TASK 4 — ERROR LOGGER*

```python
import requests
from datetime import datetime

log_file = "error_log.txt"


# ======================================================
# LOGGER FUNCTION
# ======================================================  

def log_error(function_name , error_type, message):

    '''
    Writes a structured log entry to a file.
    Standardized logs helps in debugging and analytics.
    '''

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} - {message}\n"

    try:
        # Append mode -> logs accumulate (historical tracking)
        with open(log_file, "a", encoding = "utf-8") as file:
            file.write(log_entry)


    except Exception as e:
        print("❌ failed to write log:",e)


# ======================================================
# TEST 1 - CONNECTION ERROR 
# ======================================================  

print("\n====== TESTING CONNECTION ERROR ======")

try:

    # Invalid URL -> guaranteed failure
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout = 5)

except requests.exceptions.ConnectionError as e:
    print("Connection error occurred.")

    log_error(
        function_name = "fetch_products",
        error_type = "ConnectionError",
        message = str(e)
    )

except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e))


# ======================================================
# TEST 2 - HTTP ERROR (404)
# ======================================================  

print("\n====== TESTING HTTP ERROR ======")

try:
    response.get("https://dummyjson.com/products/999", timeout = 5)

    # HTTP errors are NOT exceptions -> manual check
    if response.status_code != 200:

        message = f"{response.status_code} Not Found for product ID 999"

        print(f"HTTP error detected: {message}")

        log_error(
            function_name = "lookup_product",
            error_type = "HTTPError",
            message = message
        )
    else: 
        print("Product found successfully!")

except requests.exceptions.ConnectionError:
    log_error("lookup_product", "ConnectionError", "No internet")

except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out")

except Exception as e:
    # Adding a print here ensures you know IF the code crashed unexpectedly
    print(f"An unexpected error occured in Test 2: {e}")
    log_error("lookup_product", "UnexpectedError", str(e))


# ======================================================
# READ & DISPLAY LOG FILE
# ======================================================  

print("\n====== ERROR LOG CONTENT ======")

try:
    with open(log_file, "r", encoding = "utf-8") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    print("No logs found.")

except Exception as e:
    print("Error reading log file:", e)
```
